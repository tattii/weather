```python
%matplotlib inline 
from mpl_toolkits.basemap import Basemap 
import numpy as np
import matplotlib.pyplot as plt
import pygrib
import pandas as pd
pd.set_option('display.max_rows', 500)
```

```python
grbs = pygrib.open('Z__C_RJTD_20160220000000_GSM_GPV_Rjp_Lsurf_FD0000-0312_grib2.bin')
```

```python
def pd_grib_inventory(grbs):
    grbs.seek(0)
    df = pd.DataFrame(columns=['name', 'units' , 'level', 'valid', 'data'])
    for grb in grbs:
        df.loc[len(df) + 1] = [
            grb.name,
            grb.units,
            grb.typeOfLevel + ": " + str(grb.level),
            grb.validityDate * 10000 + grb.validityTime,
            grb.dataDate * 10000 + grb.dataTime
        ]
    return df
```

```python
pd_grib_inventory(grbs)
```
